Importing Libraries

```python

import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
```

Reading the csv file

```python

my_data = pd.read_csv('Iris.csv')
my_data
```

Gathering information about the dataset

```python

my_data.info()
```

```python

X = my_data[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']].values
X[0:5]
```

```python
Y = my_data[["Species"]]
Y[0:5]
```

Splitting the Data into Training And Testing Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_trainset, X_testset, Y_trainset, Y_testset = train_test_split(X, Y, test_size=0.3, random_state=3)
```

```python
SpeciesTree = DecisionTreeClassifier(criterion="entropy",max_depth = 4)
SpeciesTree
```

```python
SpeciesTree.fit(X_trainset,Y_trainset)
```

```python
predTree = SpeciesTree.predict(X_testset)
```

```python
print (predTree [0:5])
print (Y_testset [0:5])
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
print ("DecisionTree's Accuracy: ", metrics.accuracy_score(Y_testset,predTree))
```

Visualising the Decision Tree

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

fn=my_data.columns[1:5]
cn=my_data["Species"].unique().tolist()
SpeciesTree.fit(X, Y)
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi =300)

tree.plot_tree(SpeciesTree, feature_names =fn, class_names =cn, filled = True)
```
